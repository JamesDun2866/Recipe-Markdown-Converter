# Markdown Recipe Converter for Chowdown recipe manager. 



Simple script designed to take a recipie paste/dump and turn it into a markdown recipe format for the chowdown recpie manager. 

To use this paste a recpie into a notepad and insert the flags so the script knows where to split. These flags are TITLE, EXTRAS, DIRECTIONS And INGREDIENTS.

Source file you want to convert is opened in the first line of code change it to your text file.

Paste printed markdown into a notepad and save as a md file.

Errors: 
- Ingredients should be the last in order, the script will error if there is a newline under "INGREDIENTS" flag.
- If any of the flags are out of place script will error.
- If any flags are missing, script will error. 


Example:

Scenario: I want to convert a paper recipe from a book

- Take a photo and use Google lens to get the text of the image/scan and take the text. 
- Optional: Spellcheck and edit where required
- Paste into notepad and input flags
- Ensure file location is correct
- Run script
- Copy the markdown text and save to a new notepad file
- Save as "Recipe".md
- Upload to chowdown


####  BEFORE RUNNING, FORMAT FILE AND ADD TAGS BELOW CONTENT TITLE, EXTRAS, DIRECTIONS, INGREDIENTS (IN THAT ORDER)

```python
f = open("C:\myfile2.txt", "r", encoding="utf-8") #Load file (Example format found in repo)

Fullrec = [] #Create array for recipe
for x in f: #for each line in the recipe append to the array
    Fullrec.append(x)
try:
    for i in Fullrec: #remove extra line breaks (This creates an exception which I have handled)
        Fullrec.remove('\n')
except:
        print("")
else:
        print("")
```

```python
icount = -1 #initates a basic count 
for i in Fullrec: # for each line in fullrec append "- " to it
    icount = icount + 1
    Fullrec[icount] = "- " +Fullrec[icount]
```

```python
#Finding the title
titleindex = 0 #Initiate title index
titlearr = [] # Delcare title array 
count = 0 # declare basic count
for i in Fullrec: # for each line in the recipe append the line to titlearr until it hits the "- TITLE\n Flag"
    titlearr.append(i) 
    count = count +1
    titleindex = count
    if i ==  "- TITLE\n":
        break


titlearr.remove("- TITLE\n") # remove the flag


fulltitle = "" #declare the title varible
count = 0 #redecalre count
for x in range(len(titlearr)): #for each word in the title
    titlearr[x] = titlearr[x].lower() #lower case the entire title
    tochar = list(titlearr[x]) #Turn words into character list
    tochar.remove ('-') #Remove useless "-"
    tochar.remove (' ') #Remove useless " "
    tochar.remove ('\n') #Remove any new lines
    tochar.append(' ') #Append a space in case its a 2 word or longer recipe title
    tochar[0] =tochar[0].upper() # Upper case the first letter 
    for i in tochar: #for each character add it back to full title to stich the word back together
        fulltitle = fulltitle +i

tochar2 = list(fulltitle)#Turn the whole title into character array

del tochar2[-1] #delete the useless space
title2 = "" #declare new title
for i in tochar2: #for each character stich it back together
    title2 = title2 + i
finaltitle = title2 #set finaltitle to be used in the markdown

for i in range(titleindex): #delete the title lines in fullrec
    del Fullrec[0]
    

Extrasindex = 0 #set the extras index
Extrasarr = [] #declare the extras array
count = 0 #redeclare the count
for i in Fullrec: # for each line in fullrec add to extrasarr until it hits "- EXTRAS\n" flag
    Extrasarr.append(i)
    count = count +1
    Extrasindex = count
    if i ==  "- EXTRAS\n":
        break
        
Extrasarr.remove("- EXTRAS\n") #Remove useless flag

Extrasarr2 = [s.replace('- ', '') for s in Extrasarr] # remove all the "-" s

finalExtras = Extrasarr2 #set the final extra varible to be used in markdown

for i in range(Extrasindex): #delete the useless lines from Fullrec
    del Fullrec[0]

DirectionsIndex = 0 #Set the directions index
Directionsarr = [] #set the direcations array 
count = 0 #redeclare the count
for i in Fullrec: # for each line in Fullrec append the line to directions array untill it hits "- DIRECTIONS\n" flag
    Directionsarr.append(i)
    count = count +1
    DirectionsIndex = count
    if i ==  "- DIRECTIONS\n":
        break
        
Directionsarr.remove("- DIRECTIONS\n") #Remove useless flag

for i in range(DirectionsIndex): #remove the useless lines from fullrec
    del Fullrec[0]
    
Ingredientsarr = [] #declare ingredients array
for i in Fullrec: #for each line in Fullrec append to ingredients array untill "- INGREDIENTS" Flag
    Ingredientsarr.append(i)
    if i ==  "- INGREDIENTS":
        break
        
Ingredientsarr.remove("- INGREDIENTS") #Remove useless Flag

jpgtitle = finaltitle.replace(' ','') #To generate jpg title remove the space from finaltitle
print("Tags:")
tag = input() #Ask for tags

print("---") #Start of format
print("\n") #New Line
print("layout: recipe") #Use the recipe layout
print('title: "'+finaltitle+'"') #set the title
print("image: " +jpgtitle+".jpg")#set the jpgtitle
print("tags: " +tag )#set the tags
print("\n") # New line
print("ingredients: ") #Set the ingredient flag
for i in Ingredientsarr: #for all the ingredients print
    print(i)
print("directions: ") #set the directions flag
for i in Directionsarr: #for all the firections print
    print(i)
print("---") #end main recipe file
for i in finalExtras: #for all the extras print them
    print(i)


```
